```python
df_grouped = (
    df.groupby("age")
      .agg({
          "release_clause_eur": "mean",
          "potential": "mean"
      })
      .reset_index()
)

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_grouped,
    x="age",
    y="release_clause_eur",
    hue="potential",       
    palette="coolwarm",
    size="potential",         
    sizes=(20, 200)
)

plt.title("Age vs Release Clause (Color = Potential)", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Release Clause (€)")
plt.legend(title="Potential", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
